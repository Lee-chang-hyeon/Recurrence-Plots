```python
def rec_plot(s, eps=0.01, steps=10):
    d = pdist(s[:,None])
    d = np.floor(d/eps)
    d[d>steps] = steps
    Z = squareform(d)
    return Z
```

```python
def rec_plot1(s,eps=None, steps=None):
    if eps==None : eps=0.01
    if steps==None : steps=10
    N = s.size
    S = np.repeat(s[None,:],N, axis=0)
    Z = np.floor(np.abs(S-S.T)/eps)
    Z[Z>steps] = steps
    
    return Z
```
